```python
# solution 4:

a = int(input("angle a:"))
b = int(input("angle b:"))
c = int(input("angle c:"))
if a+b+c != 180:
    print("not a triangle")
elif (a == 90) or (b == 90) or (c == 90):
    print("Right Angle triangle") 
elif (0 < a < 90) and (0< b < 90) and (0 < c < 90):
    print("Acute")
elif (a<0) or (b<0) or (c<0):
    print("Not a Triangle")
else:
    print("Obtuse")
```

```python
# or:

# 4) Triangle Type by Angles

a = int(input("Angle a: "))
b = int(input("Angle b: "))
c = int(input("Angle c: "))

if a + b + c != 180 or min(a, b, c) <= 0:
    print("Not a triangle")         # fail-fast on invalid data
elif a == 90 or b == 90 or c == 90:
    print("Right")
elif a < 90 and b < 90 and c < 90:
    print("Acute")
else:
    print("Obtuse")
```
