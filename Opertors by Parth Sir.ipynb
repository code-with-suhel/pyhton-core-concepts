## Comparison Operators

```python
# Comparison Operators ==  !=   >    <    >=     <=
# Comparison operators allow to compare values. they give result as either True or False.

x = 78
y = 10
print(x==y)
print(x!=y)
print(x>y)
print(x<y)
print(x>=y)   # x==y or x>y
print(x<=y)   # x==y or x<y
```

# Logical Operators

```python
# Logical Operators :   and  or  not
# It is used to combine conditional statements.
# comparison and Logical Operators are most useful in Python, every machine use these operators to work(i.e work based on conditions) 
x = 13
y = 14
z = 15
print(x<y or y!=z) # True or True
```

```python
x = 18
print(x<y)
print(y!=z)
print(x<y or y!=z) # False or True
```

```python
x = 18
y = 14
z = 14
print(x<y)
print(y!=z)
print(x<y or y!=z)  # False or False
```

```python
a = 57
b = 26.5
print(a<b or a==b)  # this is same as a<=b
print(a>b or a==b)  # this is same as a>=b
```

```python
a = 57
b = 26.5
print(a<=b)
print(a>=b) 
```
