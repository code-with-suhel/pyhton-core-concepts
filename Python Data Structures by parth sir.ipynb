# Python Data structures

# 1.Lists

#### **Definition**: A List is an **Ordered, mutable collection** in python. 
- Allows Duplicates

- Elements can be of mixed data types
      
- Supports indexing(+ve & -ve) 

## Creating Lists 

```python
nums = [10,20,30]
fruits = ["mango", "apple", "banana"]
mixed = [1, "two", 3.0, [4,5]]
print("Numbers:", nums)
print("fruits:", fruits)
print("Mixed:", mixed)
```

```python
nums1 = [30,10,20] # List preserve order
print(nums1)

nums2 = [34,65,34,65]  # Duplicates Allowed
print(nums2)
```

## List Metods
- len()
- note: len() is a function not a method.

```python
fruits = ["mango", "apple", "banana"]
print("Length of fruits:", len(fruits))
```

```python
# append()
nums = [10, 20,]
nums.append(30)  # this is method not a func
print("after append:", nums)   
```

```python
# extend()
nums = [10, 20]
nums.extend([3,4,5])
print("after extend:", nums)
```

```python
# remove(value to be removed)
nums = [1,2,2,3]
nums.remove(2)
print("After Remove:", nums)
```

```python
# pop(index of the value to remove) 
nums = [10,20,30,40]
last = nums.pop()
second = nums.pop(1)
print("popped last:", last)
print("popped index 1:", second)
print("List now:", nums)
```

```python
# clear()
nums = [1,2,3]
nums.clear()
print("After Clear:", nums)
```

```python
# index()
letters = ['a','b','c','b']
print("Index of 'b':", letters.index('b'))
```

```python
# count()
letters = ['a','b','b','c','b']
print("Count of 'b':", letters.count('b'))
```

```python
# sort()
nums = [5, 2, 9, 2]
nums.sort()
print("Sorted:", nums)
```

```python
# sort()
nums = ["b","a"]
nums.sort()
print("Sorted:", nums)
```

```python
# sort(reverse)
nums = [5, 2, 9, 2]
nums.sort(reverse=True)
print("Sorted:", nums)
```

```python
# reverse()
nums = [78, 56, 89]
nums.reverse()
print("After reverse:", nums)
```

```python
# Slicing
nums = [10, 20, 30, 40, 50]
print("nums[1:4] =", nums[1:4])
print("nums[:3] =", nums[:3])
print("nums[::2] =", nums[::2])
print("nums[::-1] =", nums[::-1])
```

```python
# Concatenation & Repetition
print([1,2] + [3,4])
print([1,2] * 3)
```

# 2. Tuples

```python
# Creating Tuples
t = (10, 20, 30)
single = (42,)
print("Tuple:", t)
print("Single element tuple:", single)
```

```python
# Creating Tuples
t = (10, 20, 30)
single = (42,)   
t1 = (42)               # this is an iteger not a tuple 
print("Tuple:", t)
print(t1)
print(type(t1))
print("Single element tuple:", single)
print(type(single))
```

### Tuple Methods 

```python
# len()
# Note: len() is a function, not a method
t = (10,20,30)
print("Length:", len(t))
```

```python
# tuple.count()
t = (1,2,1,3,1)
print("Count of 1:", t.count(1))
```

```python
# tuple.index()
t = (10,20,30,20)
print("Index of 20:", t.index(20))
```

```python
# Indexing & Slicing
t = ('a','b','c','d')
print("t[0] =", t[0])
print("t[-1] =", t[-1])
print("t[1:3] =", t[1:3])
```

```python
# Concatenation & Repetition
t1 = (1,2)
t2 = (3,4)
print("t1+t2:", t1+t2)
print("t1*3:", t1*3)
print(t1*2 + t2*3)
```

```python
# Membership
t = (1,2,3)
print(2 in t)
print(5 in t)
```

# 3. Sets

### Create & Deduplicate

```python
s = {1,2,3}
dedup = set([1,2,2,3,3])   # we can also make dedup as dedup = (1,2,2,3,3)  
print("s:", s)
print("dedup:", dedup)
```

### Set Methods

```python
# len()
s = {1,1,2,3}
print("Size:", len(s))
```

```python
# add()
s = {1,2}
s.add(3)
print("After add:", s)
```

```python
s = {1,2}
print(s)
print(id(s))
```

```python
s = {1,2}
print(s)
print(id(s))
s.add(3)
print("After add:", s)
print(id(s))
```

```python
# update() : applied only for iterable data types.
# iterable data types = string, list, tuple, set, dict.
s = {1}
s.update([2,3], {4,5},(8,9))  
print("After update:", s)
```

```python
s = {1}
s.update(2)
print("After update:", s)
```

```python
s.update("hello")
s
```

```python
# remove()
s = {1,2,3}
s.remove(2)
print("After remove:", s)
```

```python
# discard()
s = {1,2}
s.discard(2)
s.discard(99)
print("After discard:", s)
```

```python
# pop()
s = {10,20,30}
removed = s.pop()
print("Removed:", removed)
print("Remaining:", s)
```

```python
# pop()
s = {40,20,30,50,90}
removed = s.pop()
print("Removed:", removed)
print("Remaining:", s)
```

```python
# clear()
s = {1,2,3}
s.clear()
print("After clear:", s)
```

```python
# union()
A, B = {1,2,3}, {3,4,5}
print("A.union(B):", A.union(B))
```

```python
# intersection()
A, B = {1,2,3}, {2,3,4,5}
print("A.intersection(B):", A.intersection(B))
```

```python
# difference()
A, B = {1,2,3}, {2,3,4,5}
print("A.difference(B):", A.difference(B))
print(B.difference(A))
```

```python
# symmetric_difference()
A, B = {1,2,3}, {2,3,4,}
print("A.symmetric_difference(B):", A.symmetric_difference(B))
print(B.symmetric_difference(A))
```

```python
# issubset, issuperset, isdisjoint
A = {1,2}
B = {1,2,3}
C = {7,8}
print("A.issubset(B):", A.issubset(B))
print("B.issuperset(A):", B.issuperset(A))
print("A.isdisjoint(C):", A.isdisjoint(C))
```

# 4.Dictionaries

### Creating Dictionaries

```python
person = {"name":"Pankaj", "role":"data scientist"}
pairs = dict([("x",1),("y",2)])
print("person:", person)
print("pairs:", pairs)
```

### Dictionary Methods

```python
# len()
d = {"a":1, "b":2}
print("Length:", len(d))
```

```python
# Access: [] vs get()
d = {"name":"Mayank", "age":28}
print("name:", d["name"])
print("salary with get:", d.get("salary"))
```

```python
# Insert/Update
d = {}
d["lang"] = "Python"
d["year"] = 1991
print("After assignments:", d)
```

```python
# update()
d = {"a":1, "b":2}
d.update({"b":20, "c":3})
print("After update:", d)
```

```python
# pop()
d = {"a":1, "b":2}
val = d.pop("a")
print("Popped a:", val, "| d now:", d)
```

```python
# popitem()
d = {"x":1, "y":2, "z":3}
item = d.popitem()
print("Popped item:", item, "| d now:", d)
```

```python
# clear()
d = {"a":1}
d.clear()
print("After clear:", d)
```

```python
# keys(), values(), items()
d = {"name":"Nishi", "role":"Data Scientist"}
print("keys:", d.keys())
print("values:", list(d.values()))
print("items:", list(d.items()))
```
