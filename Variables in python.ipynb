# Rules for Names of Variables
- A variable name must start with an alphabet or an underscore.
- A variable name cannot start with a number or any special character(@,#,$,%,^).
- A variable name can only cobtain alphabets, numbers, and underscores.
- The use of special character in prohibited.
- variable names are case sensitive.
- python keywords cannot be used as variable names.

## Important topics related to variables.
##### 1. Storing an expression in a variable 

```python
animal_1 = "c"+"a"+"t"     # "c"+"a"+"t" is an expression.
print(animal_1)
print(type(animal_1))
```

```python
animal_2 = "dog"
combined_animal = animal_1+animal_2 # animal_1 + animal_2 is an expression
print(combined_animal)
```

```python
new_variable = 12 + 34  # 12+34 is an expression
print(new_variable)
```

```python
animal_1 = "c"-"a"+"t"     # "c"+"a"+"t" is an expression.
print(animal_1)
print(type(animal_1))
```

```python
animal_1 = "c"*"a"*"t"     # "c"+"a"+"t" is an expression.
print(animal_1)
print(type(animal_1))
```

```python
animal_1 = "c"/"a"+"t"     # "c"+"a"+"t" is an expression.
print(animal_1)
print(type(animal_1))
```
